```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('TopRichestInWorld.csv')
df
```

```python
industry = df['Industry'].value_counts()
industry
```

```python
industry.index
```

```python
plt.pie(x=industry.values, labels=industry.index)
```

```python
labels = industry.index[:10]
values = industry.values[:10]

plt.pie(x=values, labels=labels, autopct='%.2f%%', colors=('blue', 'red'))
```
