```python
import pandas as pd
```

```python
df = pd.read_csv('olympics.csv', sep=';')
df
```

```python
# melt
# unpivot a dataframe, reshape dataframe wide to long
# id_vars: what we want to presereve
# value_vars: what we want to melt
df.melt(id_vars='Country', value_vars=['2021', '2022', '2023', '2024'])
```

```python
df.melt(id_vars='Country', var_name='Year', value_name='Medals')
```

```python
people = pd.read_csv('people.csv', sep=';')
people
```

```python
# pivot, reshape dataframe from long to wide
people.pivot(index='Day', columns='Name', values='Salary')
```

```python
people2 = pd.read_csv('people2.csv', sep=';')
people2
```

```python
# pivot_tanle, allows operation in dataframe, can handle duplicates
format = people2['Salary'].str.replace('€', '').str.replace(',', '').astype(int)/100
people2['Salary'] = format
people2
```

```python
people2.pivot_table(values='Salary', columns='Name', aggfunc='sum')
```

```python
# groupby, groups are more flexible than mask and do not have to create multiple masks
# groups can use operations
gender_group = people2.groupby(by='Gender')
gender_group
```

```python
gender_group.groups
```

```python
gender_group.get_group('Male').head()
```

```python
groups1 = people2.groupby(by=['Gender', 'Salary'])
groups1
```
