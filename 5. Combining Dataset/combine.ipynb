```python
import pandas as pd
```

```python
data1 = {'A':[1,2,3], 'B':[4,5,6]}
data2 = {'A':[7,8,9], 'B':[1,2,3], 'C':[8,9,8]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
# concat
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis=1)
```

```python
pd.concat([df1,df2], keys=['Data1', 'Data2'])
```

```python
pd.concat([df1,df2], keys=['Data1', 'Data2'],axis=1)
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
com1 = pd.read_csv('company21.csv', sep=';')
com2 = pd.read_csv('company22.csv', sep=';')
com1
```

```python
com2
```

```python
# merge()
# merge a datset with the database-style join
# outer join, joins everything
pd.merge(left=com1, right=com2, how='outer', on='Name')
```

```python
pd.merge(left=com1, right=com2, how='outer', on='Name', suffixes=('_21','_22'), indicator=True)
```

```python
# inner join
pd.merge(left=com1, right=com2, how='inner', on='Name', suffixes=('_21','_22'), indicator=True)
```
