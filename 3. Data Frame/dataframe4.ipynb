```python
# Broadcasting
# concept from numpy
# performs arithmetic operation between arrays or dataframe or series
data1 = pd.Series([1,2,3,4,5])
data2 = pd.Series([1,2,3,4,5])
data1 + data2
```

```python
data3 = pd.Series([1,2,3,4,5,6,7])
data1 + data3
```

```python
# conditions
male = t['sex'] == 'male'
t[male]
```

```python
t[t['age'] > 5]
```

```python
# bit wise operation & | ~
(t['sex'] == 'male') & (t['age'] > 5)
```

```python
#between
t1 = t['age'].between(20,25, inclusive='both')
t[t1]
```
