```python
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv')
df
```

```python
# apply() apply function to column and create new column
df['name_len'] = df['Name'].apply(len)
df
```

```python
del df['name_len']
df
```

```python
def strength(attack:str)->str:
    if attack > 150:
        return 'Strong'
    elif attack > 100:
        return 'Average'
    else:
        return 'Weak'
```

```python
df['Strength'] = df['Attack'].apply(strength)
df
```

```python
df.value_counts('Strength')
```

```python
# map
df['Legendary'] = df['Legendary'].map({False: 'No', True: 'Yes'})
df
```

```python
# applymap() applies function to the whole dataframe
df.applymap(type)
```

```python
# replace
# if map is used, all other values except 1 are NaN
# replace can be used in entire dataframe
df['Generation'] = df['Generation'].replace({1: 'first'})
df
```

```python
# where, returns NaN if condition is false
df.where(df['Generation'] == 2)
```

```python
# agg, aggregate multiple operations
df.agg({'Attack': ['min', 'max'], 'Defense': 'mean'})
```

```python
# copy
legendary = df['Legendary'].copy()
legendary
```

```python
# Hierarchial Indexing (multi indexing), inplace=True, chnages in original dataframe
df.set_index(['Type 1', 'Name'], inplace=True)
df
```
